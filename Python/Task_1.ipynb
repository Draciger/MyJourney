```python
# a)
# Loading the csv data in python as "df" & have a first look on the columns
df = pd.read_csv('data_PGR107.csv')
df.head()
```

```python
# b)
# Show the basic information over the numerical columns. We extract information on what the lowest (min) and highest (max) total_bill were for example.
df_description = df.describe()
print('Analysis 1:')
print(df_description)
```

```python
# b)
# An overview of the different datatypes in the raw data. We also get an overview on how many columns there are and what they are called.
print('Analysis 2:') 
d_type_analysis = df.info()
print(d_type_analysis)
```

```python
# b)
# Creating two histograms to visualize the tips and the total bills, with KDE. Further, in the analysis we go further in the data.
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
hist_tips = sns.histplot(df['tip'], bins=50, kde=True, ax=axes[0], color='green')
axes[0].set_title('Histogram Tips')
hist_bill = sns.histplot(df['total_bill'], bins=50, kde=True, ax=axes[1],color='blue')
axes[1].set_title('Histogram Total Bill')
plt.tight_layout()
print('Graphic 1:')
plt.show()
```

```python
# b)
# Boxplot´s for the different day in tips, total bills and group size.
# The circle points in the boxplot´s are for the values that are higher or lower than the quartiles.
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4)) 
sns.boxplot(x='day', y='tip', data=df, ax=axes[0], order=['Thur', 'Fri', 'Sat', 'Sun'])
axes[0].set_title('Boxplot tips')  
sns.boxplot(x='day', y='total_bill', data=df, ax=axes[1], order=['Thur', 'Fri', 'Sat', 'Sun'])
axes[1].set_title('Boxplot total bills')  
sns.boxplot(x=df['day'], y=df['size'], ax=axes[2], color='purple', order=['Thur', 'Fri', 'Sat', 'Sun'])
axes[2].set_title('Boxplot group size') 
print('Graphic 2:')
plt.show()
```

```python
#b)
# A bar-plot overview over amount of meals per day, from highest to low order.
df_value_day = df['day'].value_counts()
plt.figure(figsize=(6,4))
meals_per_day_bar = sns.barplot(x=df_value_day.index, y=df_value_day.values)
# Implement the sums
for index, value in enumerate(df_value_day.values):
    meals_per_day_bar.text(index, value, f'{value}', ha='center', va='bottom')

plt.title('Number of meals per day')
plt.ylabel('Number of meals')
plt.xlabel('day')
print('Graphic 3:')
plt.show()
```

```python
# b)
# Sum up all the bills for each day to analyze which day has the highest turnover
tot_bill_days = df.groupby(['day', 'time'])['total_bill'].sum().reset_index()

# Creating the plot
plt.figure(figsize=(6, 4))
bar_plot = sns.barplot(x='day', y='total_bill', hue='time', data=tot_bill_days, order=['Thur', 'Fri', 'Sat', 'Sun'])

# Implement the sums
for patch in bar_plot.patches:
    height = patch.get_height()
    bar_plot.annotate(f'{height:.2f}', 
                      (patch.get_x() + patch.get_width() / 2., height), 
                      ha='center', va='bottom', xytext=(0, 5), 
                      textcoords='offset points')
    
plt.title('Total Bills per day and time')
plt.ylabel('sum of total bills')
plt.xlabel('day')
plt.legend(title='Time')
print('Graphic 4:')
plt.show()
```

```python
# b)
# Analyze if there are more smokers or non-smokers that visits restaurant, during which day
smoker_sum_day = df.groupby(['day', 'smoker'])['size'].sum().reset_index()
plt.figure(figsize=(5,3))
smokers_plot = sns.barplot(x= 'day', y='size', hue='smoker', data=smoker_sum_day, order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.ylabel('sum of guests')
print('Graphic 5:')
plt.show()
```

```python
# b)
# Calculating the bill per guest for smokers and non-smokers
smoker_sum = df.groupby(['smoker'])['size'].sum().reset_index()
smoker_bill = df.groupby(['smoker'])['total_bill'].sum().reset_index()

smoker_bill['Bill_per_guest'] = smoker_bill['total_bill'] / smoker_sum['size']
print('Analysis 3:')
print(smoker_bill)

### In total the non-smokers have the highest bill but individually smokers have a higher average when it comes to per person.
```

```python
# b)
# Checking the correlation between the different categories (total_bill, tip and size)
num_columns = df.select_dtypes(include=['float', 'int']).columns
corr_columns = df[num_columns].corr()
plt.figure(figsize=(5,3))
sns.heatmap(corr_columns, annot=True, cmap='Greens')
print('Graphic 6:')
plt.show()
```

```python
# c)
# Checking if there are na values in the dataframe
na_control = df.isna().sum()
print('Analysis 4:')
print(na_control)

# Looking over the raw data, we didn't need to clean the data. Visible in analysis 4, were we checked for NA values. Since there are no instances of null values there is no need for data cleaning
```

```python
# c)
# Calculate the percentage of tips in a separate column because we only have the amount of money
# We are assuming the 'total_bill' column is the price for the food excluding the tip

df['tip_in_percentage'] = (df['tip'] * 100) / df['total_bill']
df.head()
```
