### Creating a new description field

```python
# Concatenating the 'tagline' and 'overview' fields to a new 'description' field
df['description'] = df['tagline'].str.cat(df['overview'], sep=' ')
df['description']
```

# Text Processing

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
# Function to apply text processing on a field in the dataframe
def preprocess_text(text):
    # Define the stop words
    stop_words = stopwords.words('english')  
    
    # Convert the string to lowercase
    text = text.lower() 
    
    # Remove extra white spaces
    text = re.sub(r'\s+', ' ', text)  
    
    # Remove special characters
    text = re.sub(r'[^\w\s]', '', text)
    
    # Split the string into words
    words = text.split()
    
    # Remove stop words
    words = [word for word in words if word not in stop_words]  
    
    # Stemming
    words = [PorterStemmer().stem(word) for word in words]
    
    # Lemmatizing
    words = [WordNetLemmatizer().lemmatize(word) for word in words]
    
    # Join the words back into a string
    return ' '.join(words)  

# Apply the text processing to the 'description' field
df['description'] = df['description'].apply(preprocess_text)
df['description']
```

### TF and TF-IDF representation

```python
# TF Representation of the description field
tf_vectorizer = CountVectorizer()
tf_matrix = tf_vectorizer.fit_transform(df['description'])
tf_features = tf_vectorizer.get_feature_names_out()

# Add TF representation to the DataFrame
df['tf'] = [
    {tf_features[i]: count for i, count in zip(row.indices, row.data)}
    for row in tf_matrix
]

# TF-IDF Representation of the description fields
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['description'])
tfidf_features = tfidf_vectorizer.get_feature_names_out()

# Add TF-IDF representation to the DataFrame
df['tf-idf'] = [
    {tfidf_features[i]: score for i, score in zip(row.indices, row.data)}
    for row in tfidf_matrix
]

# Display TF and TF-IDF columns
df[['tf', 'tf-idf']]
```

```python
# See the distribution of the number of words in the rows
word_counts_per_row = tf_matrix.sum(axis=1).A.flatten()  

# Calculate max, min and average number of words
max_words = word_counts_per_row.max()
min_words = word_counts_per_row.min()
average_words = word_counts_per_row.mean()

# Plot the distribution
plt.figure(figsize=(8, 5))
plt.hist(word_counts_per_row, bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Word Counts')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add statistics to the legend
stats_text = f"Max: {max_words}\nMin: {min_words}\nAverage: {average_words:.2f}"
legend_text = Line2D([0], [0], color='white', label=stats_text)

plt.legend(handles=[legend_text], loc='upper right', frameon=True, fontsize=10)

plt.show()
```

## Finding similar movies as Spider-Man

### Using cosine similarity on TF-IDF

```python
from sklearn.metrics.pairwise import cosine_similarity

# Select Spider-Man's vector
spiderman_index = df[df['title'].str.contains('Spider-Man', case=False)].index[0]
spiderman_vector = tfidf_matrix[spiderman_index]

# Compute cosine similarity with every other movie
cos_sim = cosine_similarity(spiderman_vector, tfidf_matrix).flatten()

# Add similarity scores to the DataFrame
df['similarity_with_TF-IDF'] = cos_sim

# Get top N similar movies, excluding Spider-Man
top_n = 5
similar_movies = df[df['title'] != 'Spider-Man'].sort_values(by='similarity_with_TF-IDF', ascending=False).head(top_n)

# Plot the top N similar movies and their respective Similarity Score
plt.figure(figsize=(10, 6))
plt.barh(similar_movies['title'], similar_movies['similarity_with_TF-IDF'], color='skyblue')
plt.xlabel('Similarity Score')
plt.title(f'Top {top_n} Movies Similar to Spider-Man (TF-IDF)')
plt.gca().invert_yaxis()
plt.show()
```

### Using cosine similarity with Singular Value Decomposition (SVD)

```python
from sklearn.decomposition import TruncatedSVD

# Make a svd matrix based on the TF-IDF matrix
n_components = 10
svd_model = TruncatedSVD(n_components=n_components, random_state=42)
svd_matrix = svd_model.fit_transform(tfidf_matrix)

# Locate Spider-Man's Vector
spiderman_index = df[df['title'].str.contains('Spider-Man', case=False)].index[0]
spiderman_svd_vector = svd_matrix[spiderman_index]

# Compute cosine similarity in using svd vectors
svd_sim = cosine_similarity([spiderman_svd_vector], svd_matrix).flatten()

# Add similarity scores to the DataFrame
df['similarity_with_SVD'] = svd_sim

# Get top N similar movies, excluding Spider-Man
top_n = 5
similar_movies_svd = df[df['title'] != 'Spider-Man'].sort_values(by='similarity_with_SVD', ascending=False).head(top_n)

# Plot the top N similar movies and their respective Similarity Score
plt.figure(figsize=(10, 6))
plt.barh(similar_movies_svd['title'], similar_movies_svd['similarity_with_SVD'], color='skyblue')
plt.xlabel('Similarity Score')
plt.title(f'Top {top_n} Movies Similar to Spider-Man (SVD)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
